```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
from sklearn import linear_model
import zipfile
import requests
import io
import copy, math
from sklearn.preprocessing import PolynomialFeatures
from numpy import arange
from scipy.optimize import curve_fit
from sklearn.ensemble import GradientBoostingRegressor
```

```python
# Specify the URL of the zip file
zip_file_url = "https://archive.ics.uci.edu/static/public/1/abalone.zip"

# Specify the name of the file you want to read from the zip archive
file_name_inside_zip = "abalone.data"

# Download the zip file into memory
response = requests.get(zip_file_url)
zip_bytes = io.BytesIO(response.content)

# Open the zip file
with zipfile.ZipFile(zip_bytes, 'r') as zip_ref:
    # Read the file inside the zip archive using pandas
    abalone_df = pd.read_csv(zip_ref.open(file_name_inside_zip), names=["Sex", "Length", "Diameter", "Height", "Whole weight", "Shucked weight",
                                 "Viscera weight" ,"Shell weight", "Rings"])

abalone_df.head()
```

# 3.Método de ML

Com base nos nossos dados e variáveis o melhor modelo de machine learning para aplicar com fins analiticos é a regressão linear. Uma das suas vantagens é a sua facil interpretabilidade.  Isto porque, com base nas nossas variáveis, queremos prever valores contínuos e com esta facilidade é nos fornecido uma linha que se ajusta consoante os dados para facilitar esta previsão.

```python
# Selecionar as colunas "Length" e "Whole weight" do dataframe
data = abalone_df[["Length", "Whole weight"]]

# Plotar o gráfico de dispersão
plt.scatter(data["Length"], data["Whole weight"], color="blue")


plt.xlabel("Length")
plt.ylabel("Whole weight")
plt.show()
```

Este grafico é um scatter plot antes da aplicação da normalização e da construção da regressão linear, conseguimos observar a alglomeração de pontos sendo que estes estao progressivamente a aumentar em peso a medida que o comprimento aumenta.

```python

# Criar uma instância do modelo de regressão linear
regr = linear_model.LinearRegression()

# Preparar os dados de treinamento
train_x = np.array(data[["Length"]])
train_y = np.array(data[["Whole weight"]])

# Normalizar os dados de entrada e saída
min_valx = np.min(train_x, axis=None)
max_valx = np.max(train_x, axis=None)
normalized_x = (train_x - min_valx) / (max_valx - min_valx)

min_valy = np.min(train_y, axis=None)
max_valy = np.max(train_y, axis=None)
normalized_y = (train_y - min_valy) / (max_valy - min_valy)

# Treinar o modelo de regressão linear
regr.fit(normalized_x, normalized_y)

# Coeficientes do modelo
print("coefficients:", regr.coef_)  # Slope
print("Intercept:", regr.intercept_)  # Intercept

```

Com o _coefficients_ e o _intercept_ conseguimos proceder à construção do modelo da reta:

\begin{align}
    \widehat{\text{Whole weight}} = -0.30800444 &+ 0.99021984 \cdot (\text{Length}) \nonumber \\
    \label{eq:ashenfelter}
\end{align}



```python
# Plotar o gráfico de dispersão dos dados reais e a linha de regressão linear
plt.scatter(normalized_x, normalized_y, color='blue', label='Dados reais')
# Cria um gráfico de dispersão com os dados normalizados, utilizando a cor 'blue'

plt.plot(normalized_x, regr.coef_*normalized_x + regr.intercept_, '-r', label='Regressão Linear')
# Cria uma linha de regressão linear utilizando os coeficientes do modelo treinado, utilizando a cor 'red'

plt.xlabel("Comprimento mm (Length)")
plt.ylabel("Peso total g (Whole weight)")
plt.title('Regressão Linear Simples')
plt.legend()
plt.show()
```

Na regressão linear a cima realizada conseguimos observar que quando o x toma o valor aproximadamente 0.4 existe uma grande similaridade entre os dados reais e a regresão linear, conseguimos observar também que existe regressão linear positiva devido a posição dos dados reais. E com os dados obtidos concluimos que quanto maior for o comprimento maior será o peso total.

```python

# Prever os valores de saída com base nos dados de entrada
test_y_ = regr.predict(normalized_x)

# Calcular as métricas de desempenho

# Calcula o Mean Absolute Error (MAE)
print("MAE score: %.4f" % mean_absolute_error(test_y_, normalized_y))

# Calcula o Mean Squared Error (MSE)
print("MSE score: %.4f" % mean_squared_error(test_y_, normalized_y))

# Calcula o R2-score
print("R2-score: %.4f" % r2_score(test_y_, normalized_y))

```

O MAE apresenta ótimos valores visto que quanto mais próximo do 0 é melhor e menor margem de erro tem o estudo feito. Já o MSE apresenta valores muito bons também, um modelo perfeito apresenta 0 de MSE ou seja quanto mais perto melhor e como o nosso apresenta duas casa decimais com 0 é um ótimo modelo. Falando agora no R2-score é um valor bastante bom, qualquer valor a cima de 0.7 é conciderado um valor alto havendo assim um alto nivel de correlação.

# 4.Melhorar o modelo


Apesar de termos obtido anteriormente um R2-score bastante elevado no modelo anterior decidimos tentar melhorar o nosso modelo de regressão linear. Desenvolvemos 2 modelos na tentativa de aumentar o valor de R2-score como tambem de diminuir as outras metricas (MAE e MSE)

## 4.1.Regressão Polinomial

```python
# Separando as variáveis X e y
X = abalone_df[['Length']]
y = abalone_df[['Whole weight']]

# Transformando a variável X em um polinômio de grau 2
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)

# Criando e ajustando o modelo de regressão linear polinomial
poly_regression_model = LinearRegression()
poly_regression_model.fit(X_poly, y)

# Prevendo os valores
y_pred_poly = poly_regression_model.predict(X_poly)

# Métricas de desempenho
print("MAE score: %.4f" % mean_absolute_error(y, y_pred_poly))
print("MSE score: %.4f" % mean_squared_error(y, y_pred_poly))
print("R2-score: %.4f" % r2_score(y, y_pred_poly))

# Plotando o gráfico de dispersão
plt.scatter(X, y, color='blue', label='Dados reais')
plt.plot(X, y_pred_poly, color='red', label='Regressão Polinomial')
plt.xlabel('Length')
plt.ylabel('Whole weight')
plt.title('Regressão Linear Polinomial')
plt.legend()
plt.show()
```

No modelo de regressão linear polinomial em cima apresentado podemos inferir a existencia de overfitting que é um conceito da ciência dos dados, que ocorre quando um modelo estatístico se ajusta exatamente aos seus dados de treino. Quando isto acontece, o algoritmo infelizmente não consegue ter um desempenho exato em relação a dados não vistos, anulando o seu objetivo.

## 4.2.Regressão gradient boosting

```python
# Carregando os dados de treinamento
X_train = abalone_df[["Length"]]
y_train = abalone_df[["Whole weight"]]

# Criando e treinando o modelo de regressão com Gradient Boosting
gradient_boosting_model = GradientBoostingRegressor()
gradient_boosting_model.fit(X_train, y_train)

# Fazendo previsões usando o modelo treinado
y_pred_gradient_boosting = gradient_boosting_model.predict(X_train)

# Calculando e exibindo as métricas de avaliação do modelo
print("MAE score: %.4f" % mean_absolute_error(y_train, y_pred_gradient_boosting))
print("MSE score: %.4f" % mean_squared_error(y_train, y_pred_gradient_boosting))
print("R2-score: %.4f" % r2_score(y_train, y_pred_gradient_boosting))

# Plotando o gráfico de dispersão
plt.scatter(y_train, y_pred_gradient_boosting)
plt.xlabel('Valores Reais')
plt.ylabel('Previsões (Gradient Boosting)')
plt.title('Valores Reais vs Previsões (Gradient Boosting)')
plt.plot(y_train, y_train, color='red', label='Regressão Gradient Boosting')
plt.scatter(y_train, y_pred_gradient_boosting, label='Dados', color='blue')
plt.legend()
plt.show()
```

Já neste modelo scatterplot de valores reais vs previsões conseguimos observar que existe relação linear positiva, conseguimos interpretar também que a maiorida da regressão de gradient boosting está compreendida nos dados sendo assim os valores reais e os valores previstos muito parecidos se não iguais.

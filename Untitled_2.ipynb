```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
from sklearn import linear_model
import zipfile
import requests
import io
import copy, math
from sklearn.preprocessing import PolynomialFeatures
from numpy import arange
from scipy.optimize import curve_fit
from sklearn.ensemble import GradientBoostingRegressor
```

```python
# Specify the URL of the zip file
zip_file_url = "https://archive.ics.uci.edu/static/public/1/abalone.zip"

# Specify the name of the file you want to read from the zip archive
file_name_inside_zip = "abalone.data"

# Download the zip file into memory
response = requests.get(zip_file_url)
zip_bytes = io.BytesIO(response.content)

# Open the zip file
with zipfile.ZipFile(zip_bytes, 'r') as zip_ref:
    # Read the file inside the zip archive using pandas
    abalone_df = pd.read_csv(zip_ref.open(file_name_inside_zip), names=["Sex", "Length", "Diameter", "Height", "Whole weight", "Shucked weight",
                                 "Viscera weight" ,"Shell weight", "Rings"])

abalone_df.head()
```

# 3.M√©todo de ML

Com base nos nossos dados e vari√°veis o melhor modelo de machine learning para aplicar com fins analiticos √© a regress√£o linear. Uma das suas vantagens √© a sua facil interpretabilidade.  Isto porque, com base nas nossas vari√°veis, queremos prever valores cont√≠nuos e com esta facilidade √© nos fornecido uma linha que se ajusta consoante os dados para facilitar esta previs√£o.

```python
# Selecionar as colunas "Length" e "Whole weight" do dataframe
data = abalone_df[["Length", "Whole weight"]]

# Plotar o gr√°fico de dispers√£o
plt.scatter(data["Length"], data["Whole weight"], color="blue")


plt.xlabel("Length")
plt.ylabel("Whole weight")
plt.show()
```

Este grafico √© um scatter plot antes da aplica√ß√£o da normaliza√ß√£o e da constru√ß√£o da regress√£o linear, conseguimos observar a alglomera√ß√£o de pontos sendo que estes estao progressivamente a aumentar em peso a medida que o comprimento aumenta.

```python

# Criar uma inst√¢ncia do modelo de regress√£o linear
regr = linear_model.LinearRegression()

# Preparar os dados de treinamento
train_x = np.array(data[["Length"]])
train_y = np.array(data[["Whole weight"]])

# Normalizar os dados de entrada e sa√≠da
min_valx = np.min(train_x, axis=None)
max_valx = np.max(train_x, axis=None)
normalized_x = (train_x - min_valx) / (max_valx - min_valx)

min_valy = np.min(train_y, axis=None)
max_valy = np.max(train_y, axis=None)
normalized_y = (train_y - min_valy) / (max_valy - min_valy)

# Treinar o modelo de regress√£o linear
regr.fit(normalized_x, normalized_y)

# Coeficientes do modelo
print("coefficients:", regr.coef_)  # Slope
print("Intercept:", regr.intercept_)  # Intercept

```

Com o _coefficients_ e o _intercept_ conseguimos proceder √† constru√ß√£o do modelo da reta:

\begin{align}
    \widehat{\text{Whole weight}} = -0.30800444 &+ 0.99021984 \cdot (\text{Length}) \nonumber \\
    \label{eq:ashenfelter}
\end{align}



```python
# Plotar o gr√°fico de dispers√£o dos dados reais e a linha de regress√£o linear
plt.scatter(normalized_x, normalized_y, color='blue', label='Dados reais')
# Cria um gr√°fico de dispers√£o com os dados normalizados, utilizando a cor 'blue'

plt.plot(normalized_x, regr.coef_*normalized_x + regr.intercept_, '-r', label='Regress√£o Linear')
# Cria uma linha de regress√£o linear utilizando os coeficientes do modelo treinado, utilizando a cor 'red'

plt.xlabel("Comprimento mm (Length)")
plt.ylabel("Peso total g (Whole weight)")
plt.title('Regress√£o Linear Simples')
plt.legend()
plt.show()
```

Na regress√£o linear a cima realizada conseguimos observar que quando o x toma o valor aproximadamente 0.4 existe uma grande similaridade entre os dados reais e a regres√£o linear, conseguimos observar tamb√©m que existe regress√£o linear positiva devido a posi√ß√£o dos dados reais. E com os dados obtidos concluimos que quanto maior for o comprimento maior ser√° o peso total.

```python

# Prever os valores de sa√≠da com base nos dados de entrada
test_y_ = regr.predict(normalized_x)

# Calcular as m√©tricas de desempenho

# Calcula o Mean Absolute Error (MAE)
print("MAE score: %.4f" % mean_absolute_error(test_y_, normalized_y))

# Calcula o Mean Squared Error (MSE)
print("MSE score: %.4f" % mean_squared_error(test_y_, normalized_y))

# Calcula o R2-score
print("R2-score: %.4f" % r2_score(test_y_, normalized_y))

```

O MAE apresenta √≥timos valores visto que quanto mais pr√≥ximo do 0 √© melhor e menor margem de erro tem o estudo feito. J√° o MSE apresenta valores muito bons tamb√©m, um modelo perfeito apresenta 0 de MSE ou seja quanto mais perto melhor e como o nosso apresenta duas casa decimais com 0 √© um √≥timo modelo. Falando agora no R2-score √© um valor bastante bom, qualquer valor a cima de 0.7 √© conciderado um valor alto havendo assim um alto nivel de correla√ß√£o.

# 4.Melhorar o modelo


Apesar de termos obtido anteriormente um R2-score bastante elevado no modelo anterior decidimos tentar melhorar o nosso modelo de regress√£o linear. Desenvolvemos 2 modelos na tentativa de aumentar o valor de R2-score como tambem de diminuir as outras metricas (MAE e MSE)

## 4.1.Regress√£o Polinomial

```python
# Separando as vari√°veis X e y
X = abalone_df[['Length']]
y = abalone_df[['Whole weight']]

# Transformando a vari√°vel X em um polin√¥mio de grau 2
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)

# Criando e ajustando o modelo de regress√£o linear polinomial
poly_regression_model = LinearRegression()
poly_regression_model.fit(X_poly, y)

# Prevendo os valores
y_pred_poly = poly_regression_model.predict(X_poly)

# M√©tricas de desempenho
print("MAE score: %.4f" % mean_absolute_error(y, y_pred_poly))
print("MSE score: %.4f" % mean_squared_error(y, y_pred_poly))
print("R2-score: %.4f" % r2_score(y, y_pred_poly))

# Plotando o gr√°fico de dispers√£o
plt.scatter(X, y, color='blue', label='Dados reais')
plt.plot(X, y_pred_poly, color='red', label='Regress√£o Polinomial')
plt.xlabel('Length')
plt.ylabel('Whole weight')
plt.title('Regress√£o Linear Polinomial')
plt.legend()
plt.show()
```

No modelo de regress√£o linear polinomial em cima apresentado podemos inferir a existencia de overfitting que √© um conceito da ci√™ncia dos dados, que ocorre quando um modelo estat√≠stico se ajusta exatamente aos seus dados de treino. Quando isto acontece, o algoritmo infelizmente n√£o consegue ter um desempenho exato em rela√ß√£o a dados n√£o vistos, anulando o seu objetivo.

## 4.2.Regress√£o gradient boosting

```python
# Carregando os dados de treinamento
X_train = abalone_df[["Length"]]
y_train = abalone_df[["Whole weight"]]

# Criando e treinando o modelo de regress√£o com Gradient Boosting
gradient_boosting_model = GradientBoostingRegressor()
gradient_boosting_model.fit(X_train, y_train)

# Fazendo previs√µes usando o modelo treinado
y_pred_gradient_boosting = gradient_boosting_model.predict(X_train)

# Calculando e exibindo as m√©tricas de avalia√ß√£o do modelo
print("MAE score: %.4f" % mean_absolute_error(y_train, y_pred_gradient_boosting))
print("MSE score: %.4f" % mean_squared_error(y_train, y_pred_gradient_boosting))
print("R2-score: %.4f" % r2_score(y_train, y_pred_gradient_boosting))

# Plotando o gr√°fico de dispers√£o
plt.scatter(y_train, y_pred_gradient_boosting)
plt.xlabel('Valores Reais')
plt.ylabel('Previs√µes (Gradient Boosting)')
plt.title('Valores Reais vs Previs√µes (Gradient Boosting)')
plt.plot(y_train, y_train, color='red', label='Regress√£o Gradient Boosting')
plt.scatter(y_train, y_pred_gradient_boosting, label='Dados', color='blue')
plt.legend()
plt.show()
```

J√° neste modelo scatterplot de valores reais vs previs√µes conseguimos observar que existe rela√ß√£o linear positiva, conseguimos interpretar tamb√©m que a maiorida da regress√£o de gradient boosting est√° compreendida nos dados sendo assim os valores reais e os valores previstos muito parecidos se n√£o iguais.
